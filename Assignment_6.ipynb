```python
# Survey q5: What is the effectiveness of hand hygiene in preventing health care associated infection?

import matplotlib.pyplot as plt
import numpy as np

# Data
roles = ['PGY-1', 'PGY-2', 'PGY-3']
effectiveness_levels = ['Not effective', 'Somewhat effective', 'Moderately effective', 'Highly effective', 'Very highly effective']
values = [
    [0, 0, 0, 57.2, 42.8],
    [0, 7.6, 7.6, 38.5, 46.2],
    [0, 0, 18.8, 54.1, 27.1]
]

# Convert values to NumPy array
data_array = np.array(values)

# Calculate the cumulative percentages for each role
cumulative_values = np.cumsum(data_array, axis=1)

# Create a grouped bar chart with percentage values on the bars
bar_width = 0.15
index = np.arange(len(roles))

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization

for i, effectiveness in enumerate(effectiveness_levels):
    plt.bar(index + i * bar_width, data_array[:, i], bar_width, label=effectiveness)
    for j, value in enumerate(data_array[:, i]):
        if value > 0:  # Check if the value is greater than 0
            plt.text(index[j] + i * bar_width, value + 1, f'{value:.1f}%', ha='center', va='bottom', fontsize=9, color='black')

# Adding labels and title
plt.xlabel('Roles')
plt.ylabel('Percentage')
plt.title('Effectiveness of Hand Hygiene in Preventing Healthcare-Associated Infection by Role (In Percentage)')
plt.xticks(index + bar_width * 2, roles)
plt.legend(title='Effectiveness Level')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Survey q10: What importance do your department leaders attach to the fact that you perform optimal hand hygiene?

import matplotlib.pyplot as plt
import numpy as np

# Data
roles = ['PGY-1', 'PGY-2', 'PGY-3']
importance_levels = [
    'Not important at all',
    'Slightly important',
    'Of average importance',
    'Important',
    'Very important'
]
data = [
    [0, 0, 7.1, 50.1, 42.8],
    [0, 15.3, 23.1, 30.8, 30.8],
    [0, 0, 27.1, 54.1, 18.8]
]

x = np.arange(len(roles))
height = 0.5

colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700']

fig, ax = plt.subplots(figsize=(10, 6))

bottoms = np.zeros(len(roles))

for i, level in enumerate(importance_levels):
    values = [item[i] for item in data]
    filtered_values = [val for val in values if val != 0]
    if filtered_values:
        ax.barh(x, values, height, label=level, color=colors[i], left=bottoms)
        for j, val in enumerate(values):
            if val != 0:
                ax.text(bottoms[j] + val/2, j, f'{val:.1f}%', ha='center', va='center', color='black')
        bottoms += values

ax.set_xlabel('Percentage')
ax.set_title('Importance of Optimal Hand Hygiene by Role (In Percentage)')
ax.set_yticks(x)
ax.set_yticklabels(roles)
ax.legend(title='Importance Levels', bbox_to_anchor=(1, 1))

fig.tight_layout()

plt.show()
```
